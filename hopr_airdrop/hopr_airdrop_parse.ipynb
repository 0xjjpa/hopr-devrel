```python
in_csv = 'df_1.csv'
```

```python
number_lines = sum(1 for row in (open(in_csv)))
```

```python
rowsize = 100
for i in range(1,number_lines,rowsize):
    df = pd.read_csv(in_csv,
          header=None,
          nrows = rowsize,
          skiprows = i)
    
    df = df.iloc[: , 1:]
    
    df[1] = '0' + df[1].str[1:]

    out_csv = 'input' + str(i) + '.csv'
    
    df.to_csv(out_csv,
          index=False,
          header=True,
          mode='a',
          chunksize=rowsize)
```
